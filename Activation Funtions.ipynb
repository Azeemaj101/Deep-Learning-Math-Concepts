<b>Sigmoid Function<b><br>

```python
import math
def sigmoid(x):
    return 1/(1+math.exp(-x))

sigmoid(25) #All the values between 0 to 1
```

<b>Tanh Function<b><br>

```python
def tanh(x):
    return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))

tanh(-2) #All the values between -1 to 1
```

<b>RELU Function<b><br>

```python
def RELU(x):
    return max(0,x)
RELU(-5)
```

<b>RELU Function<b><br>

```python
def RELU(x):
    return max(0.1*x,x)
RELU(-50)
```
